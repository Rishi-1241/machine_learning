```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
dataset = pd.read_csv("C:\\Users\\Prakhar Agrawal\\Downloads\\newmovies.csv")
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
main_features = ['genres' , 'keywords' , 'original_title' ,'overview' ,'tagline' , 'cast' , 'director']
```

```python
print(main_features)
```

```python
for i in main_features:
    dataset[i] = dataset[i].fillna(' ')
```

```python
combined_features = dataset['genres']+' ' + dataset['keywords']+' ' + dataset['original_title']+' ' + dataset['overview']+' ' + dataset['tagline']+' ' + dataset['cast']+' ' + dataset['director']
```

```python
print(combined_features)
```

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
similarity = cosine_similarity(feature_vectors)
```

```python
similarity.shape
```

```python
print(similarity)
```

```python
movie_name = input(' Enter your favourite movie name : ')
```

```python
title_list = dataset['title'].tolist()
```

```python
print(title_list)
```

```python
close_match = difflib.get_close_matches(movie_name, title_list)
```

```python
print(close_match)
```

```python
closest = close_match[0]
```

```python
print(closest)
```

```python
index = dataset[dataset.title==closest]['index'].values[0]
```

```python
print(index)
```

```python
similarity_score = list(enumerate(similarity[index]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print(sorted_similar_movies)
```

```python
print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = dataset[dataset.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```

```python
movie_name = input(' Enter your favourite movie name : ')

title_list = dataset['title'].tolist()

close_match = difflib.get_close_matches(movie_name, title_list)

closest = close_match[0]

index = dataset[dataset.title == closest]['index'].values[0]

similarity_score = list(enumerate(similarity[index]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = dataset[dataset.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
