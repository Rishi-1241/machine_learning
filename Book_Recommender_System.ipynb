## COLLABRATIVE BASED RECCONMENDER SYSTEM

```python
new = ratings_with_name.groupby('User-ID').count()['Book-Rating'] >200
```

```python
padhe_likhe = new[new].index
```

```python
padhe_likhe
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe)]
```

```python
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import difflib
```

```python
title_list = books['Book-Title'].tolist()
```

```python
book_name = input('enter the book name : ')
```

```python
close_match = difflib.get_close_matches(book_name, title_list)
```

```python
close_match
```

```python
closest = close_match[0]
```

```python
closest
```

```python
similarity = cosine_similarity(pt)
```

```python
similarity
```

```python
similarity.shape
```

```python
index = np.where(pt.index=='Harry Potter')
```

```python
index
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity[index])),key=lambda x:x[1],reverse=True)[0:5]
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return(data)
```

```python
book = input('enter the book name : ')
```

```python
recommend(book)
```
